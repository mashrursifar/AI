```python
import torch
```

```python
# requires_grad is a boolean attribute of a tensor that controls whether PyTorch should 
# track operations on that tensor to compute gradients automatically.

w = torch.tensor(2.0, requires_grad=True) 
x = torch.tensor(4.0)
y = w * x
print(y)
```

```python
print(y.grad)
```

```python
# computation graph -> hidden graph of how many numbers are calculated. It happens automatically
#  
w.grad.zero_()
x = torch.tensor(2.0)
y = torch.tensor(4.0, requires_grad=True)
y_predict = w * x
y_true = torch.tensor(10.0)

loss = (y_true - y_predict)**2
print(loss)

loss.backward() # how much each learnable number caused the mistake. only take one backward 
# w.grand_zero() reset
print(w.grad) # stores instruuction so that we know how to change the number so that low mistake. it wil tell direction and amount. 

''' w.grad >0  -> decrease
    w.grad <0  -> inrease
    new_value = old_vlue - learning_rate * gradiant'''
```

```python
''' If 'w.grad' negative then w should increase
    and viceversa'''

```

```python
# input
x = torch.tensor(2.0)

# Weight
w = torch.tensor(3.0, requires_grad=True)

# predicted
y_pred = w * x

# expected result
y_true = torch.tensor(10.0)
loss = (y_pred - y_true) **2
lr = 0.1
loss.backward()

print("Before Update")
print("W: ",w.item())
print("loss: ",loss.item())
print("gradient: ",w.grad.item())

with torch.no_grad(): # temporarily change behaivour and then go back to normal
    w -= lr * w.grad 

w.grad.zero_()

print("After Update")
print("W: ",w.item())
print("loss: ",loss)
```
